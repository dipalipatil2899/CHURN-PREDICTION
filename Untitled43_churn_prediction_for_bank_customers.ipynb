```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
i=data_df.loc[:,["EstimatedSalary","Balance"]]
i
```

```python

d=data_df.loc[:,["CreditScore"]]
d
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(i,d,test_size=0.30)
```

```python
print(xtest)
```

```python
print(ytest)
```

```python
from sklearn.linear_model import LinearRegression
lnr=LinearRegression()
lnr.fit(xtrain,ytrain)
```

```python
predicted_creditscore=lnr.predict(xtest)
predicted_creditscore
```

```python
plt.scatter(xtest["EstimatedSalary"],ytest,color='b')
plt.plot(xtest,predicted_creditscore,color='r',linestyle="-")
plt.legend("best")
plt.title("Linear Regression")
plt.xlabel("EstimatedSalary")
plt.ylabel("CreditScore")
```

```python

plt.scatter(xtest["Balance"],ytest,color='y')
plt.plot(xtest,predicted_creditscore,color='g',linestyle="-")
plt.legend("best")
plt.title("Linear Regression")
plt.xlabel("balance")
plt.ylabel("creditscore")
```
